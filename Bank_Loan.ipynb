```python
cat_feats = ['purpose']
```

```python
import pandas as pd
```

```python
final_data = pd.DataFrame()
def tree():
    final_data = pd.get_dummies(loans, columns=cat_feats, drop_first=True)
    return final_data
final_data = tree()
   
```

```python
X = pd.DataFrame()
def tree():
    X = final_data.drop('not.fully.paid',axis=1)
    return X
X = tree()
```

```python
y = pd.DataFrame()
def tree():
    y = final_data.loc[:,final_data.columns == 'not.fully.paid']
    return y
y = tree()
```

```python
from sklearn.model_selection import train_test_split

def tree():
     return train_test_split(X, y, test_size=0.30, random_state=101)
X_train, X_test, y_train, y_test = tree()
```

```python
def tree():
    print(X_train.shape)
    print(y_train.shape)
tree()
```

```python
def tree():
    print(X_test.shape)
    print(y_test.shape)
tree()
```

```python
from sklearn.tree import DecisionTreeClassifier
def tree():
    dtree = DecisionTreeClassifier()
    return dtree
dtree = tree()
```

```python
def tree():
    dtree.fit(X_train,y_train)
    
tree()
```

```python
predictions = pd.DataFrame()
def tree():
    predictions = dtree.predict(X_test) 
    return predictions
predictions = tree()    
```

```python
from sklearn.metrics import confusion_matrix
def tree():
    print(confusion_matrix(y_test,predictions))
tree()
```

```python
from sklearn.metrics import recall_score
def log():
    recall2 = recall_score(y_test,predictions)
    print('Recall score for test data is:', recall2)
    return recall2
recall2 = log()
```

```python
from sklearn.metrics import classification_report
def tree():
    print(classification_report(y_test,predictions))
tree()
```
